5. Perform meaningful visualizations using the dataset. Bring at least 3
meaningful visualizations.


```python


data = {
   'SL': np.random.normal(loc=5.5, scale=0.8, size=150),
    'SW': np.random.normal(loc=3.0, scale=0.4, size=150),
    'PL': np.random.normal(loc=4.5, scale=1.0, size=150),
    'PW': np.random.normal(loc=1.3, scale=0.5, size=150),
   'Classification': np.random.choice(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'], size=150)
}

df = pd.DataFrame(data)


sns.pairplot(df, hue='Classification', markers=["o", "s", "D"])
plt.suptitle('Pairplot of Iris Features by Classification', y=1.02)
plt.show()


plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['SL', 'SW', 'PL', 'PW']], hue=df['Classification'])
plt.title('Boxplot of Iris Sepal and Petal Measurements')
plt.show()


plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='PL', y='PW', hue='Classification', style='Classification', markers=["o", "s", "D"])
plt.title('Petal Length vs Petal Width by Classification')
plt.show()
```
